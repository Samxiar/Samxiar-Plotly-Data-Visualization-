```python
import plotly.express as px
```

```python
px.pie(df,names='Population(2011)')
```

```python
px.histogram(df, x = 'Population(2001)', nbins = 2)
```

```python
px.box(df,
       y= 'Population(2001)',
       title='Population(2001)',
       hover_data= ['Population(2001)']
)
```

```python
fig = px.scatter(df,x='Population(2011)',y='Population(2001)')
```

```python
fig.show()
```

```python
px.bar(df, x='Population(2001)', y='Population(2011)', color ='City')
```
